```python
# SCRAP TITLE AND CITATION NUMBER IN LIST
title_table = []
cit_table = []
n1 = 0
n2 = 0
for table_link in topic_links:
  n2 += 1
  req = requests.get(f"{base_url}{table_link}")
  webpage = BeautifulSoup(req.content)
  links = [a.get("href") for a in webpage.findAll("a", {"class":"LinkButton"})]
  for link in links:
    re = requests.get(f"{base_url}{link}")
    site = BeautifulSoup(re.content)
    title = site.find('h1', {'style':"text-align: center"}).contents[0]
    if len(site.find('div', style = "color:orange").contents) <= 1:
      citation = 0
    elif len(site.find('div', style = "color:orange").contents) == 4:
      citation = str(site.find('div', style = "color:orange").contents[2].split(':')[1]).strip()
    elif len(site.find('div', style = "color:orange").contents) == 2:
      citation = str(site.find('div', style = "color:orange").contents[0].split(':')[1]).strip()
    title_table.append(title)
    cit_table.append(citation)
    n1 += 1
    print(f'link {n1} from topic {n2}')
```

```python
#CREATE TABLE FOR ARTICLE TITLE AND CITATION
import pandas as pd
table = {'Title': title_table, 'Citation':cit_table}
df = pd.DataFrame(table)
df['Citation'] = df['Citation'].astype('int64')
df.sort_values(by=['Citation'], inplace = True)
df #print the dataframe
```

## **QUESTION 2**

```python
# SCRAP LINKS IN LIST FOR DETAILS (LDO NAME, NUMBER OF LDOs, TITLE, YEAR, CITATION)
LDO_table = []
LDO_name_table = []
year_table = []
title_table = []
cit_table = []
n1 = 0
n2 = 0
for table_link in topic_links:
  n2 += 1
  req = requests.get(f"{base_url}{table_link}")
  webpage = BeautifulSoup(req.content)
  links = [a.get("href") for a in webpage.findAll("a", {"class":"LinkButton"})]
  for link in links:
    re = requests.get(f"{base_url}{link}")
    site = BeautifulSoup(re.content)
    title = site.find('title').contents[0]
    if len(site.find('div', style = "color:orange").contents) <= 1:
      citation = 0
    elif len(site.find('div', style = "color:orange").contents) == 4:
      citation = str(site.find('div', style = "color:orange").contents[2].split(':')[1]).strip()
    elif len(site.find('div', style = "color:orange").contents) == 2:
      citation = str(site.find('div', style = "color:orange").contents[0].split(':')[1]).strip()
    LDO = len(site.find('h2', text = "Links, Downloads and Online Resources").find_next_sibling().findAll('div')) - 1 #exclude parent div
    year = str(site.find('h3', text = 'BibTeX').find_next_sibling()).split(',')
    for yea in year:
      if yea.find('year') != -1:
        date = yea.split('{')[1][:-1]
    LDO_name = site.find('h2', text = "Links, Downloads and Online Resources").find_next_sibling().findAll('a')
    names_list = []
    for name in LDO_name:
      if name.getText() not in ("", " "):
        names_list.append(name.getText())
    title_table.append(title)
    cit_table.append(citation)
    LDO_table.append(LDO)
    year_table.append(date)
    LDO_name_table.append(names_list)
    n1 += 1
    print(f'link {n1} from topic {n2}')
```

```python
# create dataframe with scraped results and define the column datatype
import pandas as pd
table = {'Title': title_table, 'Citation':cit_table, 'LDO Names':LDO_name_table ,'Number of LDOs':LDO_table, 'Year': year_table}
df = pd.DataFrame(table)
df['Citation'] = df['Citation'].astype('int64')
df['Year'] = df['Year'].astype('int64')
df.sort_values(by=['Citation'], inplace = True)
df
```

```python
# TOP 25 CITED ARTICLES
top_25 = df.nlargest(25,['Number of LDOs'])
top_25
```

```python
# CREATE A PLOT TABLE
plot_table = df[['Number of LDOs',	'Year']].groupby('Year').mean()
plot_table
```

```python
re = requests.get("https://sitescrape.awh.durham.ac.uk/comp42315/pbl_tnsre2022features4670.htm")
site = BeautifulSoup(re.content)
arthur_link = site.find('div', style = "margin-left: var(--size-marginleft);text-align: center").findAll('a')
arthur_names = [a.getText() for a in arthur_link]
arthur_names
```

```python
# SCRAP LINKS IN LIST FOR DETAILS
impact_factor = []
topic_table = []
arthur_table = []
title_table = []
cit_table = []
LDO_table = []
LDO_name_table = []
n1 = 0
n2 = 0
for text in list(topic_text.keys()):
  n2 += 1
  topic = text
  topic_link = topic_text[f"{text}"]
  req = requests.get(f"{base_url}{topic_link}")
  webpage = BeautifulSoup(req.content)
  links = [a.get("href") for a in webpage.findAll("a", {"class":"LinkButton"})]
  for link in links:
    re = requests.get(f"{base_url}{link}")
    site = BeautifulSoup(re.content)
    title = site.find('h1', {'style':"text-align: center"}).contents[0]
    if len(site.find('div', style = "color:orange").contents) <= 1:
      citation = 0
      impact = 0
    elif len(site.find('div', style = "color:orange").contents) == 4:
      citation = str(site.find('div', style = "color:orange").contents[2].split(':')[1]).strip()
      impact = str(site.find('div', style = "color:orange").contents[0].split(':')[1]).strip()
    elif len(site.find('div', style = "color:orange").contents) == 3:
      citation = 0
      impact = str(site.find('div', style = "color:orange").contents[0].split(':')[1]).strip()
    elif len(site.find('div', style = "color:orange").contents) == 2:
      citation = str(site.find('div', style = "color:orange").contents[0].split(':')[1]).strip()
      impact = 0
    arthur_link = site.find('div', style = "margin-left: var(--size-marginleft);text-align: center").findAll('a')
    arthur_names = [a.getText() for a in arthur_link]
    LDO = len(site.find('h2', text = "Links, Downloads and Online Resources").find_next_sibling().findAll('div')) - 1 #exclude parent div
    LDO_name = site.find('h2', text = "Links, Downloads and Online Resources").find_next_sibling().findAll('a')
    names_list = []
    for name in LDO_name:
      if name.getText() not in ("", " "):
        names_list.append(name.getText())
    for arthur in arthur_names:
      impact_factor.append(impact)
      topic_table.append(topic)
      arthur_table.append(arthur)
      title_table.append(title)
      cit_table.append(citation)
      LDO_table.append(LDO)
      LDO_name_table.append(names_list)
    n1 += 1
    print(f'link {n1} from topic {n2}')
```

```python
# create dataframe with scraped results and define the column datatype
import pandas as pd
table = {'Title': title_table, 'Citation':cit_table,'Research Topic': topic_table , 'Arthurs Names': arthur_table, 'Inpact Factor': impact_factor, 'LDO Names':LDO_name_table ,'Number of LDOs':LDO_table}
df = pd.DataFrame(table)
df['Citation'] = df['Citation'].astype('int64')
df['Number of LDOs'] = df['Number of LDOs'].astype('int64')
df
```
